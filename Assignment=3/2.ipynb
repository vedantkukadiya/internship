```python
class Solution:
    def Soldire(self, rating):
        n = len(rating)
        tmp = 0
        for j in range(n):
            l_sml = l_lrg = 0
            r_sml = r_lrg = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    l_sml += 1
                else:
                    l_lrg += 1

            for k in range(j+1, n):
                if rating[k] > rating[j]:
                    r_lrg += 1
                else:
                    r_sml += 1

            tmp += l_sml * r_lrg
            
            tmp += l_lrg * r_sml

        return tmp
s = Solution()
print(s.Soldire([2,5,3,4,1]))
```
