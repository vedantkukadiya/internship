```python
class Solution:
    def Student(self, startTime, endTime, queryTime):
        count = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1
        return count
startTime = [1, 2, 3]
endTime = [3, 2, 7]
queryTime = 4

s=Solution()
print(s.Student(startTime, endTime, queryTime))
```
