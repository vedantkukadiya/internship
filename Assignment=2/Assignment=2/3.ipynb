```python
class Solution:
    def numIdenticalPairs(self, nums: list[int]) -> int:
        res = 0
        
        for i in range(len(nums)):
          for j in range(len(nums)):
            if nums[i] == nums[j] and i < j:
              res += 1
              
        return res
s = Solution()
print(s.numIdenticalPairs([1,2,1,2,3,1]))
```
