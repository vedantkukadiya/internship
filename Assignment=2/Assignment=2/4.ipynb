```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: list[int]) -> list[int]:
        tmp = []
        for i in nums:
            c = 0
            for j in nums:
                if j < i:
                    c += 1
            tmp.append(c)
        return tmp

s = Solution()
print(s.smallerNumbersThanCurrent([8,1,2,2,3]))
```
